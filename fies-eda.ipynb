# **Introduction**


This notebook explores the Family Income and Expenditure Survey (FIES) 2012 Vol 1 provided by the Philippine Statistics Authority. This survey was designed to provide income and expenditure data for 17 regions in the Philippines and was used as the basis for the policies and programs for the government to improve the socio-economic conditions of the Filipinos.

## **Sampling and Data Collection**

**Unit of Analysis**

The statistics gathered from the survey were based on the characteristics of private households. For this study, a family in a private household is defined as the household head, spouse, unmarried children, ever-married children, son-in-law/daughter-in-law, parents of the head/spouse and other relatives who are members of the household. Additionally, two or more persons not related to each other by blood, marriage or adoption are also considered in this survey. A notable insight however is that  only the income and expenditure of the member who is considered as the household head are included.


**Implications of the Data Collection Method**

Though the participants of the survey were carefully monitored during the time of taking it, some implications regarding the validity of the data would need to be questioned as due to the large number of information needed to complete the survey and the lack of information regarding the survey process provided by the PSA, there could have been instances that families could have fabricated answers to speed up the survey-taking process.

**Dataset Usage**

As the dataset came from the Philippine Statistics Authority, the FIES serves as a reliable source that can be used for studies regarding studies involving household information, expenditures, and income classifications.

# **Data Cleaning**

To prepare for exploratory data analysis, we opted to convert the variables of interest (with a corresponding categorical equivalent) to their categorical equivalent based on the library provided with the dataset.


We will be using three libraries in this Notebook.


*   `numpy` provides comprehensive mathematical functions
*   `pandas` offers data structures and functions for data analysis
*   `matplotlib` allows data to be rendered and visualized



```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

To begin, we must load the dataset using `pandas`. The `read_csv()` function will load the dataset found in the path provided. We assign this dataframe to the `raw_fies_df` variable.

```python
raw_fies_df = pd.read_csv("FIES PUF 2012 Vol.1.CSV")
```

To see if the data was loaded correctly, we can call the `info()` function on `raw_fies_df` to view general information about the dataframe. To see all the columns in our large dataset, we need to pass the parameter `verbose=True`.

```python
raw_fies_df.info(verbose=True)
```

Next, we select only the columns that will be used for exploratory data analysis. Afterwards, we can call the `head()` function to view the dataframe.

```python
fies_df = raw_fies_df[['URB', 'FSIZE','KB', 'JOB',
                       'TOINC', 'NATDC',
                       'T_HEALTH', 'T_TOTEX']]
fies_df.head()
```

There are multiple things we need to fix on this dataset. First, there are missing values in the `KB` column. Since all other values in `KB` are integers, we can call pandas' `to_numeric()` function to convert the entire column to a numeric data type. We also pass the parameter `errors='coerce'` which converts non-numeric values, such as the empty values, to `NA`. Next, we call the `fillna()` function to convert all `NA` values to `0`. Finally, since some columns may be converted to a float, we cast it all to integer through `astype()`.

```python
fies_df = fies_df.apply(pd.to_numeric, errors='coerce').fillna(0).astype(int)
fies_df.head()
```

To make the table more readable, we can convert the keys in the `URB`, `JOB` and `KB` columns to their actual values. The `replace()` function can easily do this so long as we can specify specific key-value pairs.

```python
fies_df = fies_df.replace({
    'URB': {
        1: "URBAN",
        2: "RURAL"
    },
    'JOB': {
        1: "EMPLOYED",
        2: "UNEMPLOYED"
    },
})
```

However, since categories in `KB` are represented by a range of values, we need a helper function to determine the job industry. We then apply this function to the `KB` column as shown below.

```python
def get_job_industry(key):
  if 111 <= key <= 329:
    job_industry = "Agriculture, Forestry and Fishing"
  elif 510 <= key <= 990:
    job_industry = "Mining and quarrying"
  elif 1011 <= key <= 3320:
    job_industry = "Manufacturing"
  elif 3510 <= key <= 3530:
    job_industry = "Electricity, Gas, Steam and Airconditioning supply"
  elif 3600 <= key <= 3900:
    job_industry = "Water supply, sewerage, waste management and Remediation Activities"
  elif 4100 <= key <= 4390:
    job_industry = "Construction"
  elif 4510 <= key <= 4799:
    job_industry = "Wholesale and retail trade; Repair of motor vehicles and motorcycles"
  elif 4911 <= key <= 5320:
    job_industry = "Transportation and Storage"
  elif 5510 <= key <= 5630:
    job_industry = "Accommodation and Food Service Activities"
  elif 5811 <= key <= 6399:
    job_industry = "Information and communication"
  elif 6411 <= key <= 6630:
    job_industry = "Financial and insurance activities"
  elif 6811 <= key <= 6820:
    job_industry = "Real estate activities"
  elif 6910 <= key <= 7500:
    job_industry = "Professional, Scientific and Technical activities"
  elif 7710 <= key <= 8299:
    job_industry = "Administrative and Support Service activities"
  elif 8411 <= key <= 8430:
    job_industry = "Public Administration and defense; compulsory social security"
  elif 8511 <= key <= 8560:
    job_industry = "Education"
  elif 8611 <= key <= 8890:
    job_industry = "Human health and social work services"
  elif 9000 <= key <= 9329:
    job_industry = "Arts, entertainment and recreation"
  elif 9411 <= key <= 9690:
    job_industry = "Other service activities"
  elif 9700 == key:
    job_industry = "Activities of households as employers of domestic personnel"
  elif 9901 <= key <= 9909:
    job_industry = "Activities of extra-territorial organizations and bodies"
  else:
    job_industry = "Unknown"

  return job_industry

fies_df['KB'] = fies_df['KB'].apply(get_job_industry)
```

# Exploratory Data Analysis

## **Question #1: Is there a correlation between healthcare expenditure and income decile?**


The first step is to separate the data into different groups based on their national income decile. This is done through the `groupby()` function, to which we pass the parameter `NATDC` for National Income Decile. After grouping, the `agg()` function calculates aggregate values for the `T_HEALTH` or Total Healthcare Expenditure column. In this case, we calculated the mean, median, and standard deviation of each decile. These values are then rounded to two decimal places for clarity.


```python
healthcare_expenditure_across_deciles = fies_df.groupby('NATDC').agg({'T_HEALTH': ['mean', 'median', 'std']}).round(2)


healthcare_expenditure_across_deciles
```

Regarding healthcare expenditure, it may be more meaningful to look at the median compared to the mean because its distribution is heavily skewed to the right. To demonstrate, we can simply view the histogram through matplotlib's `hist()` function.

```python
plt.hist(fies_df['T_HEALTH'], bins=100)
plt.xlabel('Healthcare Expenditure')
plt.ylabel('Frequency')
plt.show()
```

Finally, to visualize how the median healthcare expenditure varies across income deciles, we can use the simple bar graph from matplotlib's `.bar()` function. The graph will consist of the income deciles and median healthcare expenditure for the x-axis and y-axis, respectively.

```python
plt.bar(healthcare_expenditure_across_deciles['T_HEALTH']['median'].index, healthcare_expenditure_across_deciles['T_HEALTH']['median'].values)
plt.xlabel('Income Decile')
plt.ylabel('Median Healthcare Expenditure')
plt.title('Median Healthcare Expenditure by Income Decile')
plt.xticks(healthcare_expenditure_across_deciles['T_HEALTH']['median'].index)
plt.show()
```

Both the numerical summary and visualization suggest that there may be a positive relationship between healthcare expenditure and income decile. However, explicitly defining a relationship is a another task.

In order to quantify the correlation between healthcare expenditure and income decile, we can use Spearman's Correlation. Spearman's Correlation was specifically chosen because income decile is ordinal by nature. Furthermore, healthcare expenditure data is heavily skewed and does not have a normal distribution. Thus, Pearson's Correlation would not have been appropriate.

To perform this calculation, we must first slice the dataframe to only include the relevant columns: `T_HEALTH` and `NATDC`. Then, we use the `corr()` function from the pandas library where we specify the correlation method (`spearman`). This returns a pairwise matrix, from which we access only the correlation between `T_HEALTH` and `NATDC`.

```python
spearman_corr = fies_df[['T_HEALTH', 'NATDC']].corr(method='spearman')['T_HEALTH']['NATDC']
print(f'Spearman Correlation between Health Expenditure and Income Decile = {spearman_corr.round(2)}')
```

The resulting correlation coefficiant tells us two things. First, it is a positive relationship, suggesting that belonging to a higher income decile correlates to having a higher healthcare expenditure. Second, the strength of the relationship is 0.49, which can be classified as moderate to high.

## **Question #2: Is there a correlation between healthcare expenditure and total expenditure in households with non-zero healthcare expenditure?**

For this question, we are interested in whether or not a higher total expenditure is correlated with a higher healthcare expenditure as well.

As previously stated, health expenditure data is heavily skewed to the right. The same is true for the total expenditure. However, since we are only interested in the relationship between the two variables, and not the values themselves, we can perform a log transformation in order to achieve a normal distribution.

In order to perform the log transformation, we use `apply()`  which takes a function as a paremeter and applies that function to every entry in the series. In this case, we use numpy's `log1p()`. This function calculates `log(1 + x)`, which is necessary because of the presence of 0 values in the data.

```python
plt.hist(fies_df['T_HEALTH'], bins=100)
plt.title('Distribution of Healthcare Expenditure BEFORE log transformation')
plt.xlabel('Healthcare Expenditure')
plt.ylabel('Frequency')
plt.show()

plt.hist(fies_df['T_HEALTH'].apply(np.log1p), bins=100)
plt.title('Distribution of Healthcare Expenditure AFTER log transformation')
plt.xlabel('Healthcare Expenditure')
plt.ylabel('Frequency')
plt.show()
```

The same transformation is done with the total expenditure column.

```python
plt.hist(fies_df['T_TOTEX'], bins=100)
plt.title('Distribution of Total Expenditure BEFORE log transformation')
plt.xlabel('Total Expenditure')
plt.ylabel('Frequency')
plt.show()

plt.hist(fies_df['T_TOTEX'].apply(np.log1p), bins=100)
plt.title('Distribution of Total Expenditure AFTER log transformation')
plt.xlabel('Total Expenditure')
plt.ylabel('Frequency')
plt.show()
```

Next, we can plot these two variables in a scatterplot through matplotlib's `scatter()`.

```python

log_data = fies_df[['T_HEALTH', 'T_TOTEX']].apply(np.log1p)
plt.scatter(log_data['T_TOTEX'], log_data['T_HEALTH'])
plt.xlabel('Total Expenditure')
plt.ylabel('Healthcare Expenditure')
plt.show()
```

The scatterplot reveals two interesting shapes. First, we see that there is a positive correlation between total expenditure and healthcare expenditure. However, there is a significant number of points which are located at `y=0`, forming a straight line at the bottom of the graph. These data points show that across nearly all expenditure levels, there exists households who do not spend money on healthcare. With this in mind, it may be more meaningful to exclude this group of households (those who have 0 healthcare expenditure) when analyzing the correlation.

To remove the households with 0 healthcare expenditure, we simply filter the rows in the `fies_df` to have greater than 0 `T_HEALTH`.

```python
non_zero_t_health = fies_df[fies_df['T_HEALTH'] > 0]

log_data = non_zero_t_health[['T_HEALTH', 'T_TOTEX']].apply(np.log1p)
plt.scatter(log_data['T_TOTEX'], log_data['T_HEALTH'])
plt.xlabel('Total Expenditure')
plt.ylabel('Healthcare Expenditure')
plt.title('Healthcare Expenditure vs Total Expenditure')
plt.show()
```

We will use the same function `corr()` as in the previous question to calculate the pearson correlation.

```python
log_data.corr(method='pearson')
```

The results show that there is a moderate positive correlation between healthcare expenditure and total expenditure.

# **Clustering**

To help answer this research question, we can employ a data modelling technique known as *cluster analysis* or *clustering*. Cluster analysis is the act of identifying groups of objects within datasets, where objects in the same cluster are more alike to each other than to objects of different clusters (Nvidia, n.d.). A well-known method to create clusters from a dataset is through the *K-Means algorithm*, which we will be using.

## **Pre-Processing**

The K-Means algorithm works by iteratively partitioning a dataset into similar groups based on their euclidian distance to a set of centroids. Its heavy reliance on space and distance brings challenges when working on datasets with high-dimensionality. To avoid this issue, our cluster analysis will be based solely on two columns: Total Income *(TOINC)* and Total Health Expenditure *(T_HEALTH)*. To begin, let us visualize these two variables on a scatter plot.

```python
plt.scatter(fies_df['TOINC'], fies_df['T_HEALTH'])
plt.show()
```

The scatterplot reveals the presence of some extreme outliers on both axes. This may be problematic because the clusters will be formed based on the structure of data. Thus, we will remove the outliers by dropping all data points that are more than 3 standard deviations away from the mean.

To do this, we will calculate the z-score through the `z-score` function from `scipy`'s `stats` module. Then, we calculate its absolute value using `numpy`'s `abs` function, and ensure it is less than 3 for every row in the income and health expenditure columns.

```python
from scipy import stats

trimmed_df = fies_df[(np.abs(stats.zscore(fies_df[['TOINC', 'T_HEALTH']])) < 3).all(axis=1)]
```

```python
plt.scatter(trimmed_df['TOINC'], trimmed_df['T_HEALTH'])
plt.show()
```

Now, we can see that the data no longer contains outliers.

Next, we need to eliminate any differences in scale between the two features using `StandardScaler`. The `fit_transform()` function will return a scaled copy of any data we pass.

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_data = scaler.fit_transform(trimmed_df[['TOINC', 'T_HEALTH']])
scaled_df = pd.DataFrame(scaled_data,
                         columns=['TOINC', 'T_HEALTH'])
scaled_df.head()
```

## **Data Modelling**

The K-Means algorithm requires us to specify a number of clusters for the dataset. To choose the most optimal, we can use a visual approach known as the *elbow method*. This approach involves calculating how well the dataset "fits" into different numbers of clusters based on some evaluation metric. In this case, we will calculate the *inertia*, which is the sum of squared distance between each point and the centroid of their cluster. We will repeatedly calculate this value over an increasing number of clusters, and visualize it on a plot.

Using `sklearn`'s `KMeans` module to perform clustering, we simply need to specify the number of clusters and pass our data to the `fit()` function. We will do this repeatedly from 1-20 and store the inertia in an array. Then, we use `matplotlib`'s `plot()` function to visualize the inertias.

```python
from sklearn.cluster import KMeans

inertias = []

for i in range(1,21):
    kmeans = KMeans(n_clusters=i, random_state=30)
    kmeans.fit(scaled_df)
    inertias.append(kmeans.inertia_)

plt.plot(range(1,21), inertias, marker='o')
plt.title('Elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.xticks(np.arange(0, 21, 1))
plt.show()
```

As the name suggests, we will select the number of clusters which correspond to the "elbow" of the line. Through visual inspection, we can determine that the most optimal number of clusters is 5. Thus, we will specify `n_clusters = 5` and fit our data. We also pass a second parameter `random_state=30`. This ensures that our results are reproducible.

```python
kmeans = KMeans(n_clusters=5, random_state=30)
kmeans.fit(scaled_df)
```

We will then use the `copy()` function to copy the trimmed dataset with untransformed features and add a new column called *CLUSTER* which will hold the calculated cluster labels from the `labels_` attribute.

```python
clustered_df = trimmed_df.copy()
clustered_df['CLUSTER'] = kmeans.labels_
clustered_df.head()
```

Finally, we can visualize the clusters using `scatter()` from `matplotlib`. We will also use `get_cmap()` function to get 5 colors from the `rainbow` colormap, one for each cluster.

```python
cmap = plt.get_cmap('rainbow', 5)
centroids = kmeans.cluster_centers_
centroids_original = scaler.inverse_transform(centroids)

plt.figure(figsize=(10,7))
plt.scatter(clustered_df['TOINC'],
            clustered_df['T_HEALTH'],
            c=clustered_df['CLUSTER'],
            cmap=cmap)
plt.xlabel('Total Income')
plt.ylabel('Total Health Expenditure')
plt.title('Clusters')

plt.colorbar(ticks=range(5))
plt.show()
```

# **Hypothesis Testing**

In order to determine which factors significantly affect the household expenditure allocated to healthcare, we opted to perform hypothesis testing on the most relevant variables with regards to healthcare expenditure based on discussion and the exploratory data analysis, more specifically, we aimed to answer the following questions.


**Questions:**
1. Is there a significant difference in the mean household expenditure allocated to healthcare between those from urban and rural households?
2. Is there a significant difference in the mean household expenditure allocated to healthcare between families whose head had a job or business in the last 6 months?
3. Is there a significant difference in the mean household expenditure allocated to healthcare depending on the type of industry of the household head?
4. Is there a significant difference in the mean household expenditure allocated to healthcare across different family sizes?

## **Question #1: Is there a significant difference in the mean household expenditure allocated to healthcare between those from urban and rural households?**

In order to determine whether there is a significant difference in the mean between the two household classifications, we can perform Welch's t-test. Welch's t-test is the appropriate test for determininng whether there is a significant difference between the mean of two populations if both the sample size and variance greatly varies. Considering that this test is susceptible to outliers, we will first perform outlier detection by retaining all values 3 standard deviations away from the mean.

```python
from scipy import stats
from scipy.stats import ttest_ind

#Slicing the dataframe to get the urban and rural columns and their respective total health expenditure
health_expenditure_urban = fies_df[fies_df['URB'] == 'URBAN']['T_HEALTH']
health_expenditure_rural = fies_df[fies_df['URB'] == 'RURAL']['T_HEALTH']

#Removing outliers
trimmed_urban_df = health_expenditure_urban[(np.abs(stats.zscore(health_expenditure_urban)) < 3)]
trimmed_rural_df = health_expenditure_rural[(np.abs(stats.zscore(health_expenditure_rural)) < 3)]
```

After cleaning the data, we can now perform hypothesis testing and determine whether to reject or accept the null hypothesis.

$H_0$ (null hypothesis): There is no significant difference among the mean total healthcare expenditure of the two household type classifications

$H_A$ (alternative hypothesis): There is a significant difference among the mean total healthcare expenditure of the two household type classifications

```python
#Performing t-test
t_statistic, p_value = ttest_ind(trimmed_urban_df, trimmed_rural_df, equal_var=False)

print('T-statistic: {:.2f}'.format(t_statistic))
print(f'P-value: {p_value}')
```

Given that the P-value is incredibly small and much much smaller than 0.05, we have sufficient evidence to reject the null hypothesis and conclude that **There is a significant difference among the mean total healthcare expenditure of the two classifications**

**Observation:** This is likely due to the difference of healthcare environments between urban and rural communities.

## **Question #2: Is there a significant difference in the mean household expenditure allocated to healthcare between families whose heads had a job or not in the last 6 months?**

In order to determine whether there is a significant difference in the mean household expenditure between households whose head of the family had a job or not in the last 6 months, we can also perform a t-test following the same conditions as the previous question, leading us to determine whether to reject or accept the null hypothesis.

$H_0$ (null hypothesis): There is no significant difference among the mean total healthcare expenditure of the two classifications (Job and No Job)

$H_A$ (alternative hypothesis): There is a significant difference among the mean total healthcare expenditure of the two classifications (Job and No Job)

```python
#Slicing the dataframe to get the columns that indicate whether they have or not have had a job in the last 6 months and their respective total health expenditure
health_expenditure_job = fies_df[fies_df['JOB'] == 'EMPLOYED']['T_HEALTH']
health_expenditure_no_job = fies_df[fies_df['JOB'] == 'UNEMPLOYED']['T_HEALTH']

#Removing outliers
trimmed_job_df = health_expenditure_job[(np.abs(stats.zscore(health_expenditure_job))<3)]
trimmed_no_job_df = health_expenditure_no_job[(np.abs(stats.zscore(health_expenditure_no_job))<3)]

#Performing t-test
t_statistic, p_value = ttest_ind(trimmed_job_df, trimmed_no_job_df, equal_var=False)

print('T-statistic: {:.2f}'.format(t_statistic))
print(f'P-value: {p_value}')
```

Given that the p-value is incredibly small and much much smaller than 0.05, we have sufficient evidence to reject the null hypothesis and conclude **There is a significant difference among the mean total healthcare expenditure of the two classifications (Job and No Job)**

**Observation:** This is likely due to how the spending power of the family would be affected if they were employed or not in the last 6 months.

## **Question#3: Is there a significant difference in the mean household expenditure allocated to healthcare depending on the type of industry of the household head?**

This question aims to tackle whether that assumption is valid and to do this, we will be using one-way ANOVA to determine whether a group with a significantly different mean healthcare expenditure exists.

First, we will remove those with Unknown as their industry considering this question focuses on known kinds of business. Next we will clean the data by removing outliers as these greatly affect the results of ANOVA testing.
